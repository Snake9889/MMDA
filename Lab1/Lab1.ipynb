```python
from scipy import linalg
import matplotlib.pyplot as plt
import matplotlib as mpl
from matplotlib import colors
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.model_selection import train_test_split
```

```python
iris = datasets.load_iris()
X = iris.data[:, :2]  # we only take the first two features.
Y = iris.target

def LDA(classifier):
    x_min, x_max = X[:, 0].min() - .5, X[:, 0].max() + .5
    y_min, y_max = X[:, 1].min() - .5, X[:, 1].max() + .5
    h = .02  # step size in the mesh
    xx, yy = np.meshgrid(np.arange(x_min, x_max, h), np.arange(y_min, y_max, h))
    Z = classifier.predict(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)
    plt.figure(1, figsize=(10, 8))
    plt.pcolormesh(xx, yy, Z, cmap=plt.cm.Spectral)
    plt.scatter(X[:, 0], X[:, 1], c=Y, edgecolors='k', cmap=plt.cm.Paired)
    plt.xlabel('Sepal length')
    plt.ylabel('Sepal width')
    plt.xlim(xx.min(), xx.max())
    plt.ylim(yy.min(), yy.max())
    plt.xticks(())
    plt.yticks(())
    plt.show()

plt.title('Linear Discriminant Analysis')
lda = LinearDiscriminantAnalysis(solver="svd", store_covariance=True).fit(X, Y)
LDA(lda) 
```

```python
from sklearn.discriminant_analysis import QuadraticDiscriminantAnalysis

X_ = iris.data[:, :2]  # we only take the first two features.
Y_ = iris.target
X = X_[Y_ != 0]
Y = Y_[Y_ != 0]
qda = QuadraticDiscriminantAnalysis(store_covariance=True).fit(X, Y)
plt.title('Quadratic Discriminant Analysis')
LDA(qda) 
```

```python
from sklearn.linear_model import LogisticRegression

logreg = LogisticRegression(C=1e5).fit(X, Y)
plt.title('Logistic Regression')
LDA(logreg)
```

```python
from sklearn import svm

clf = svm.SVC(kernel="linear").fit(X, Y)
plt.title('SVM (линейное ядро).')
LDA(clf)
```

```python
clf = svm.SVC(degree=2).fit(X, Y)
plt.title('SVM (квадратичное ядро).')
LDA(clf)
```
